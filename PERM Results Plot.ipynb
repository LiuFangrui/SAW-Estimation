```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Extract data

```python
data = pd.read_csv('perm data.csv')
data
```

# Calculate $\mu$

```python
data['mu'] = np.power(data['cL'], 1/data['L'])
data
```

# Plot

```python
# Create plot with dots for each point
plt.plot(data['L'], data['mu'], 'o-', label='μ(L)')

# Add average line (dotted)
avg_mu = data['mu'].mean()
plt.axhline(y=avg_mu, color='r', linestyle='--', label=f'Average μ = {avg_mu:.2f}')

# Annotate last point
last_L = data['L'].iloc[-1]
last_mu = data['mu'].iloc[-1]
plt.annotate(f'({last_L}, {last_mu:.2f})', 
             xy=(last_L, last_mu),
             xytext=(10, 10), 
             textcoords='offset points',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
             arrowprops=dict(arrowstyle='->'))

# Add labels and title
plt.xlabel('L (Walk Length)')
plt.ylabel('μ')
plt.title('How μ Changes with Walk Length L using PERM')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
